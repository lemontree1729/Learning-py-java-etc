```python
import requests
from bs4 import BeautifulSoup
import os

print(os.getcwd())
```

```python
origin = requests.get("http://www.naver.com")
html = origin.text
print(html)
```

```python
soup = BeautifulSoup(html, "html.parser")
print(soup)
```

```python
word = soup.select("#NM_set_home_btn")
print(word)
```

```python
print(word[0].attrs)
```

```python
today = soup.select("#NM_THEME_CONTAINER p, em, strong")
for i in today:
    print(i.text)
```

```python
samsung = BeautifulSoup(
    requests.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EC%82%BC%EC%84%B1").text,
    "html.parser",
)
for i in samsung.select(".news_tit"):
    print(i.text, i.attrs["href"])
```

```python
user = BeautifulSoup(requests.get(f"https://search.naver.com/search.naver?where=news&sm=tab_jum&query={input()}").text)
for i in user.select(".news_tit"):
    print(i.text, i.attrs["href"])
```

```python
movie = BeautifulSoup(requests.get(f"https://movie.naver.com/movie/sdb/rank/rmovie.naver").text)
for index, tag in enumerate(movie.select(".title>div>a")):
    print(f"{index+1}. {tag.text} => {tag.attrs['href']}")
```

```python
import pyautogui
base_url = "https://search.naver.com/search.naver"
search_word = pyautogui.prompt('enter the word')
search_condition = f"?where=news&sm=tab_jum&query={search_word}"
cnt=0
while search_condition:
    if cnt == 10:
        break
    cnt+=1
    url_parse = BeautifulSoup(requests.get(base_url + search_condition).text)
    for i in url_parse.select(".news_tit"):
        print(i.text, i.attrs["href"])
    search_condition = url_parse.select_one("a.btn_next").attrs["href"]
```

```python
search_word = pyautogui.prompt('enter the word')
for i in range(10):
    url_parse = BeautifulSoup(requests.get(f"https://search.naver.com/search.naver?where=news&sm=tab_jum&query={search_word}&start={10*i+1}").text)
    for i in url_parse.select(".news_tit"):
        print(i.text, i.attrs["href"])
```
