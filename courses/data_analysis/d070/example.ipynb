```python
import requests
from bs4 import BeautifulSoup
```

```python
origin = requests.get("https://movie.naver.com/movie/running/current.naver")
parse_origin = BeautifulSoup(origin.text)
movies = list(map(lambda x: x.text, parse_origin.select("dt.tit>a")))
stars = list(map(lambda x: float(x.text), parse_origin.select("dl.info_star span.num")))
for i in range(10):
    print(f"{i+1}. {movies[i]} / 별점: {stars[i]}")
```

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/Sol.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
barlist = plt.bar(x=movies[:10], height=stars[:10])
maxindex = stars.index(max(stars[:10]))
barlist[maxindex].set_color('r')
plt.xlabel("movies")
plt.xticks(rotation=90)
plt.ylabel("stars")
plt.grid(True, axis='y', color='black',alpha=0.5, linestyle='--')
plt.text(maxindex, stars[maxindex], stars[maxindex], c='blue', ha="center")
plt.show()
```
