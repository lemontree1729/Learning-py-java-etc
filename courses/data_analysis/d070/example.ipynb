```python
import requests
from bs4 import BeautifulSoup
```

```python
origin = requests.get("https://movie.naver.com/movie/running/current.naver")
parse_origin = BeautifulSoup(origin.text)
movies = list(map(lambda x: x.text, parse_origin.select("dt.tit>a")))
stars = list(map(lambda x: float(x.text), parse_origin.select("dl.info_star span.num")))
for i in range(10):
    print(f"{i+1}. {movies[i]} / 별점: {stars[i]}")
```

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/Sol.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
barlist = plt.bar(x=movies[:10], height=stars[:10])
print(*barlist)
barlist[max(enumerate(barlist), key = lambda x:x[1].get_height())[0]].set_color('r')
plt.xlabel("movies")
plt.xticks(rotation=45)
plt.ylabel("star")
plt.show()
```
