```python
from bs4 import BeautifulSoup
import requests
url = "http://www.yjjob.or.kr/p/?j=23"
```

```python
req = requests.get(url)
soup = BeautifulSoup(req.text)
```

```python
import re

datas = list(map(lambda x:x.text,soup.select("div.r-eview")))
wanted_datas = []
for data in datas:
    pre_data = list(map(lambda x:re.sub("\t+|^ +| +$", "", x), re.split(" : |\n", data)))
    wanted_data = {"과정구분":pre_data[1], "과정명":pre_data[2]}
    cnt = 3
    while(cnt+1 < len(wanted_data)):
        wanted_data[pre_data[cnt]] = pre_data[cnt+1]
        cnt += 2
    wanted_data.pop("")
    wanted_datas.append(wanted_data)
print(*wanted_datas, sep="\n")
```
